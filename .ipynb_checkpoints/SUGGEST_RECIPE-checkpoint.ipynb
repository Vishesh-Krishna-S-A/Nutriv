```python
import pandas as pd
import random
```

```python
recipes = pd.read_csv('Nutriv.csv')
```

```python
recipes
```

```python
def filter_by_calories(df, min_calories, max_calories):
    return df[(df['calories'] >= min_calories) & (df['calories'] <= max_calories) & (df['breakfast'] >= 1)]
```

```python
filtered_recipes = filter_by_calories(recipes, 400, 600)
```

```python
filtered_recipes
```

```python
filtered_recipes.loc[:, 'title':'breakfast']
```

```python
def filter_by_cal(df, min_calories, max_calories):
    return df[(df['calories'] >= min_calories) & (df['calories'] <= max_calories) & (df['breakfast'] >= 1)]

filter_C =filter_by_cal(recipes, 400, 600)
recommended_recipe = filter_C.sample(n=2)

# Step 5: Print out the recommended recipes
print(recommended_recipe.loc[:, 'title':'breakfast'])
```

```python
def filter_by_cal2(df, min_calories, max_calories):
    return df[(df['calories'] >= min_calories) & (df['calories'] <= max_calories) &( (df['lunch']>=1) | df['dinner']>=1 | (df['breakfast']>=1))]

filter_C =filter_by_cal2(recipes, 900, 1200)
filter_C = filter_C.loc[:,'title':'lunch']
```

```python
recommended_recipe = filter_C.sample(n=2)

recommended_recipe
```
